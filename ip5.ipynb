```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
invalidation = pd.read_csv('DirectApplications.csv')
```

```python
invalidation
```

```python
invalidation.columns
```

```python
country_group = invalidation.groupby('CountryOffice ')
country_group
```

```python
invalidation['log_fillings'] = np.log(invalidation['Fillings'])
```

```python
invalidation['log10_grants'] = np.log10(invalidation['Grants'])
```

```python

country_group = invalidation.groupby('CountryOffice ')
country_group

for country, data in country_group:
    plt.plot(data['Year'], data['log10_grants'], label=country)

# Add labels and a legend
plt.xlabel('Year')
plt.ylabel('log10 Grants')
plt.legend()

# Display the plot
plt.show()
```

```python
invalidation['log10_fillings'] = np.log10(invalidation['Fillings'])
```

```python

country_group = invalidation.groupby('CountryOffice ')
country_group

for country, data in country_group:
    plt.plot(data['Year'], data['Grants'], label=country)

# Add labels and a legend
plt.xlabel('Year')
plt.ylabel('Grants')
plt.legend()

# Display the plot
plt.show()
```

```python
for country, data in country_group:
    plt.plot(data['Year'], data['Fillings'], label=country)

# Add labels and a legend
plt.xlabel('Year')
plt.ylabel('Fillings')
plt.legend()

# Display the plot
plt.show()
```

```python

for country, data in country_group:
    plt.plot(data['Year'], data['Invalidation'], label=country)

# Add labels and a legend
plt.xlabel('Year')
plt.ylabel('Invalidation')
plt.legend()

# Add point markers to mark observations
for country, data in country_group:
    plt.scatter(data['Year'], data['Invalidation'], label=f'{country} - Observation', s=20, marker='o')


# Customize x-axis ticks to show more distinctions
plt.xticks(invalidation['Year'].unique(),rotation = 90)

# Display the plot
plt.show()
```

```python
uspto = pd.read_csv('UsptoGrantsIPC2021.csv')

uspto.describe()
```

```python
uspto10 = uspto.sort_values(by = 'Total', ascending=False).head(10)

uspto10.plot( x = 'IPC', y = 'Total', kind = 'bar')
plt.show()
```

```python
cnipa5 = cnipa10.head(5)
cnipa5['Country'] = ['China' for i in range(5)]
cnipa5
```

```python
uspto5 = uspto10.head(5)
uspto5['Country'] = ['US' for i in range(5)]
uspto5
```

```python
epo5 = epo10.head(5)
epo5['Country'] = ['Europe' for i in range(5)]
epo5
```

```python
kipo5 = kipo10.head(5)
kipo5['Country'] = ['Korea' for i in range(5)]
kipo5
```

```python
jpo5 = jpo10.head(5)
jpo5['Country'] = ['Japan' for i in range(5)]
jpo5
```

```python
merged_df = pd.concat([uspto5,jpo5,kipo5,epo5,cnipa5],ignore_index=True)
merged_df
```

```python
# Pivot the data to create a DataFrame suitable for a stacked bar chart
pivot_df = merged_df.pivot(index='IPC', columns='Country', values='Total')

# Plot a stacked bar chart
ax = pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6))

# Customize the chart
plt.xlabel('IPC')
plt.ylabel('Total')
plt.title('Stacked Bar Chart by IPC and Country')

plt.show()
```

```python
utility = pd.read_csv('examiner-utility-patent.csv')
utility.describe()
```

```python
utility1 = utility.iloc[:,0:3]
```

```python


# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
utility1.plot(kind='bar', stacked=True, ax=ax)

# Adding labels and title
plt.xlabel('Examiner')
plt.ylabel('Count')
plt.title('Stacked Bar Chart of Examiner Data')

# Show the plot
plt.show()
```

```python

df_percentage = utility.divide(utility['Total '], axis=0) * 100

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
df_percentage.drop('Total ', axis=1).plot(kind='bar', stacked=True, ax=ax)

# Adding labels and title
plt.xlabel('Examiner')
plt.ylabel('Percentage')
plt.title('Stacked Bar Chart as Percentage of Total')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
weighted = pd.read_csv('WSV_country_opinion_weighted_variables.csv')

weighted = weighted.iloc[:,1:]
weighted
```

```python
idv = pd.read_csv('collective_index_country_level.csv')

idv = idv.iloc[:,1:]
idv
```

```python
merged_df = idv.merge(weighted,on='Code',how = 'outer')
merged_df = merged_df[~merged_df.weighted_Q188.isna()]
merged_df = merged_df[~merged_df.idv.isna()]
```

```python
merged_df
```

```python
# Plot the data
plt.figure(figsize=(10, 6))

# Scatter plot for normalized_idv and weighted_Q188
plt.scatter(merged_df["normalized_idv"], merged_df["weighted_Q188"])

# Add country labels
for i, row in merged_df.iterrows():
    plt.annotate(row["Code"], (row["normalized_idv"], row["weighted_Q188"]), textcoords="offset points", xytext=(0, 10))

# Labels and legend
plt.xlabel("normalized_idv")
plt.ylabel("weighted_Q188")
plt.legend(loc="upper right")

# Show the plot
plt.show()
```

```python
gdp = pd.read_csv('gdp_countries.csv',skiprows=4)
```

```python
gdp = gdp[['Country Name','Country Code','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']]

gdp['Average_gdp'] = gdp[['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']].mean(axis = 1)
gdp
```

```python
gdp.drop(columns=['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022'],inplace=True)
```

```python
gdp.to_csv('final_average_gdp.csv')
```

```python
average_gdp = pd.read_csv('final_average_gdp.csv')
```

```python
average_gdp = average_gdp[['Country Code','Average_gdp']]
average_gdp.rename(columns={'Country Code' : 'Code'}, inplace=True)
average_gdp
```

```python
merged_df = merged_df.merge(average_gdp,on='Code',how='outer')
```

```python
merged_df = merged_df[~merged_df.normalized_idv.isna()]
```

```python
merged_df
```

```python
merged_df = merged_df[~merged_df['Average_gdp'].isna()]
```

```python
merged_df
```

```python
sm.graphics.plot_partregress(data = merged_df,endog = 'weighted_Q197',exog_i = 'normalized_idv',exog_others=['Average_gdp'],obs_labels=False)
plt.show()
```

```python

prim = yale_data[['varname','GeoName','2021','2020','2019','2018','2017','2016','2015','2014','2013','2012','2011','2010','2008']]
prim['Code'] = prim['GeoName'].map(state_alpha_codes)
```

```python
prim
```

#congress-Estimated percentage who think Congress should be doing more/much more to address global warming						

```python
congress = prim[prim.varname == 'congress']
congress = congress[~congress['2021'].isna()]
congress 
```

```python
cols_for_calculation = congress.columns[2:15]
cols_for_calculation
```

```python
mean2 = congress[['2021','2020']].mean(axis =1)
mean2
```

```python


# Calculate the standard deviation
std_dev = congress[cols_for_calculation].std(axis=1)

# Calculate the mean
mean_value = congress[cols_for_calculation].mean(axis=1)

congress['variation'] = std_dev/mean_value
```

```python
congress.drop(columns=cols_for_calculation,inplace=True)
congress
```

```python
count_above_threshold = len(congress[congress['variation'] > 0.1])


print(f'number of states with a variaton > 0.1: {count_above_threshold}')
```

```python
plt.figure(figsize=(10, 6))
plt.bar(congress['Code'], congress['variation'])
plt.xlabel('Code')
plt.ylabel('Variation of congress')
plt.title('Variation by Code')
plt.xticks(rotation=90)
plt.axhline(y=0.06, color='red', linestyle='--', label='Threshold at 0.06')
plt.show()
```

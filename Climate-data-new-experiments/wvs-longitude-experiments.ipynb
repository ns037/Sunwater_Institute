```python
longitudinal = pd.read_csv('WVS_TimeSeries_4_0.csv')
```

```python
longitudinal_I002 = longitudinal[['COUNTRY_ALPHA','S017','I002']]
```

```python
longitudinal_I002.describe()
```

```python
longitudinal_E217 = longitudinal[['COUNTRY_ALPHA','S017','E217']]
longitudinal_A197 = longitudinal[['COUNTRY_ALPHA','S017','A197']]
```

```python
longitudinal_A197.describe()
```

```python
longitudinal_E217.describe()
```

```python

longitudinal_E217 = longitudinal_E217[longitudinal_E217.E217 > 0]
longitudinal_E217.E217 = longitudinal_E217.E217.apply(lambda x: 1 if x == 10 else 0)
longitudinal_E217['weighted_E217'] = longitudinal_E217.E217 * longitudinal_E217.S017
longitudinal_E217.drop(columns=['E217','S017'],inplace=True)
longitudinal_E217 =longitudinal_E217.groupby('COUNTRY_ALPHA')['weighted_E217'].mean().reset_index()
longitudinal_E217
```

```python
longitudinal_I002 = longitudinal_I002[longitudinal_I002.I002 > 0]
longitudinal_I002.I002 = longitudinal_I002.I002.apply(lambda x: 1 if x == 1 else 0)
longitudinal_I002['weighted_I002'] = longitudinal_I002.I002 * longitudinal_I002.S017
longitudinal_I002.drop(columns=['I002','S017'],inplace=True)
longitudinal_I002 =longitudinal_I002.groupby('COUNTRY_ALPHA')['weighted_I002'].mean().reset_index()
longitudinal_I002
```

```python
longitudinal_E217.to_csv('longitudinal_E217.csv')
```

```python
longitudinal_E217 = pd.read_csv('longitudinal_E217.csv')
longitudinal_I002 = pd.read_csv('longitudinal_I002.csv')
```

```python
longitudinal_E217.drop(columns='Unnamed: 0',inplace=True)
```

```python
longitudinal_I002.drop(columns='Unnamed: 0',inplace=True)
```

```python
longitudinal_E217.rename(columns={'COUNTRY_ALPHA':'Code'},inplace=True)
longitudinal_I002.rename(columns={'COUNTRY_ALPHA':'Code'},inplace=True)
```

```python
merged_df = idv.merge(longitudinal_E217,on = 'Code',how = 'outer')
```

```python
merged_df=merged_df[(~merged_df.normalized_idv.isna())& (~merged_df.weighted_E217.isna())]
merged_df
```

```python
for i, row in merged_df.iterrows():
    plt.scatter(row['normalized_idv'], row['weighted_E217'], label=row['Code'],s = 100)
    plt.text(row['normalized_idv'], row['weighted_E217'], row['Code'], fontsize=6, ha='right', va='bottom')

# Label axes
plt.xlabel('normalized_idv')
plt.ylabel('weighted_E217')



# Show the plot
plt.grid()
```

```python
merged_df = idv.merge(longitudinal_I002,on = 'Code',how = 'outer')
merged_df=merged_df[(~merged_df.normalized_idv.isna())& (~merged_df.weighted_I002.isna())]

for i, row in merged_df.iterrows():
    plt.scatter(row['normalized_idv'], row['weighted_I002'], label=row['Code'],s = 100)
    plt.text(row['normalized_idv'], row['weighted_I002'], row['Code'], fontsize=6, ha='right', va='bottom')

# Label axes
plt.xlabel('normalized_idv')
plt.ylabel('weighted_I002')



# Show the plot
plt.grid()
```
